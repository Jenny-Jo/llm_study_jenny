```python
import re
from pytube import Playlist
url_playlist = "https://www.youtube.com/watch?v=gdm322JcuAM&list=PLj_F-FxRi9JX2X2NO_JI69-WZ-izMFFXZ"
# url_playlist = "https://www.youtube.com/watch?v=gJQZQ7Oowsc&list=PLj_F-FxRi9JWb5ZScK0zy0a8M0rIPCQzC"
# url_playlist="https://www.youtube.com/watch?v=I-GIBkCNFmQ&list=PLPP1JwaE5jQZLwOaw04FeBI_G3ATeVOSG"
playlist = Playlist(url_playlist)
```

```python
playlist._video_regex = re.compile(r"\"url\":\"(/watch\?v=[\w-]*)")
```

```python
len(playlist.video_urls)
```

```python
playlist.video_urls[0]
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api.formatters import SRTFormatter, TextFormatter

def get_video_id(video_url):
    video_id = video_url.split('v=')[1][:11]
    return video_id
```

```python
video_id = get_video_id(playlist.video_urls[0])
video_id
```

```python
transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=['ko'])
" ".join(line['text'] for line in transcript)
```

# youtube playlist

```python
from youtube import get_youtube_video_info,get_video_urls

url_playlist="https://www.youtube.com/watch?v=gJQZQ7Oowsc&list=PLj_F-FxRi9JWb5ZScK0zy0a8M0rIPCQzC"
video_urls=get_video_urls(url_playlist)
len(video_urls)
```

```python
video_info=get_youtube_video_info(video_urls[0])
```

```python
video_info
```

```python
# url, title, channel, caption -> 파일로 저장
import json
info = {
    "url": video_urls[0],
    "title": video_info['title'],
    "channel": video_info['channel'],
    "caption": video_info['caption']
}

with open('video_info.json', 'w', encoding='utf-8') as f:
    json.dump(info, f, ensure_ascii=False, indent=4)


```

```python
# 플레이리스트 전체를 json 파일로 저장
import json

video_info=get_youtube_video_info(video_urls[0])
video_info_list=[]
for i in video_urls:
    video_info=get_youtube_video_info(i)
    info = {
        "url": i,
        "title": video_info['title'],
        "channel": video_info['channel'],
        "caption": video_info['caption']
    }
    video_info_list.append(info)
with open('video_info.json', 'w', encoding='utf-8') as f:
    json.dump(video_info_list, f, ensure_ascii=False, indent=4)



```
